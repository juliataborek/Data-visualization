```python
f = open("espanol.txt")
text = f.read().split()
f.close()
```

```python
slowa = list(text)
```

```python
slowa = list(filter(lambda x: x.replace('subj', ''), slowa))
```

```python
slowa[1][0]+slowa[1][1]
```

```python
slowa1 = []
for i in range(len(slowa)):
    if len(slowa[i])>=2:
        slowa1.append(slowa[i][0]+slowa[i][1])
```

```python
c = Counter(slowa1)
```

```python
c1=list(c.most_common())
a1 = []
for i in range(0, len(c1)):
    a1.append(c1[i][0])
a2 = []
for i in range(0, len(c1)):
    a2.append(c1[i][1])
```

```python
c1
```

```python
unique_string=(" ").join(a1)
frequency = a2
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10,ranks_only= frequency).generate(unique_string)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
```
