```python
from matplotlib import image
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sympy
```

```python
r = [0.3870993*1.49*10**11,0.723336*1.49*10**11,1.000003*1.49*10**11,1.52371*1.49*10**11,5.2029*1.49*10**11,9.537*1.49*10**11,
     19.189*1.49*10**11,
     30.0699*1.49*10**11] 
T = [0.2408467*31557600,0.61519726*31557600,1.0000174*31557600,1.8808158*31557600,11.862615*31557600,29.447498*31557600,
     84.016846*31557600,164.79132*31557600]
```

```python
def getImage(paths):
    return OffsetImage(plt.imread(path))

paths = [
    'merkury.png',
    'wenus.png',
    'ziemia.png',
    'mars.png',
    'jowisz.png',
    'saturn.png',
    'uran.png',
    'neptun.png']

plt.style.use('dark_background')
fig = plt.figure()
ax = fig.add_subplot()


fig.set_figheight(10)
fig.set_figwidth(10)

ax.plot(T,r,color='green')

ax.set_title('Trzecie prawo Keplera' )
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')


ax.set_ylim(5*10**10,10**13)
ax.set_yscale("log")
ax.set_xlim(6*10**6,10**10)
ax.set_xscale("log")

a = r'\frac{T^3}{r^2} = const'
plt.annotate('$%s$' %a, xy = (2*10**7,3*10**12),size = 15 ).set_color('white')
plt.ylabel('półoś wielka (r)')
plt.xlabel('okres orbitalny (T)')

ax.scatter(T, r)


for x0, y0, path in zip(T,r,paths):
    ab = AnnotationBbox(getImage(paths), (x0, y0), frameon=False)
    ax.add_artist(ab)

    
plt.savefig('03_Tab_272348.pdf')
plt.show()
```
