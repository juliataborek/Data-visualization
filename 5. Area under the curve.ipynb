```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sympy
import random
from matplotlib.patches import Rectangle
```

```python
x = np.arange(0,5,0.1)
y = 1/ (1+x**3)
```

```python
ilosc = random.randint(1,30)
```

```python
x1 = np.linspace(0,5,ilosc+1)
```

```python
x1
```

```python
y1 = []
for i in range(0,ilosc+1):
    y1.append(1/(1+ x1[i]**3))
```

```python
x2=[]
for i in range(1, ilosc+1):
    x2.append(x1[i])

    
y2 = []
for i in range(0,ilosc):
    y2.append(1/(1+ x2[i]**3))
```

```python
def integral_approximation(f, a, b):
    return (b-a)*np.mean(f)

def f1(x):
    return 1/(1+ x**3)

a = 0
b = 5

x_range = np.arange(a,b+0.0001,.0001)
fx = f1(x_range)

pole = integral_approximation(fx,a,b)
pole
```

```python
pole = (5-0)*np.mean(y)
pole
```

```python
plt.figure(figsize=(16,6))

ax1=plt.subplot(1,3,1)
ax1.set_xlim(left= 0)
ax1.set_ylim(bottom = 0)
ax1.set_xlim(0,5)
ax1.plot(x,y)
ax1.scatter(x1,y1, zorder = 5)
pole1 = 0
for i in range(0, ilosc):
    ax1.add_patch(Rectangle((x1[i], 0), x1[1], y1[i], facecolor="lightblue", edgecolor="darkblue"))
    pole1 = pole1 + y1[i]*x1[1]
plt.title(round(pole1,3))


ax2=plt.subplot(1,3,2)
ax2.set_xlim(left= 0)
ax2.set_ylim(bottom = 0)
ax2.set_xlim(0,5)
ax2.plot(x, y)
plt.title('Przedziałów jest ' + str (ilosc) + ', pole = ' + str(round(pole,3))  )
ax2.fill_between(x, y, color='lightblue')
a = r'f(x) = \frac{1}{1+ x^3}'
plt.annotate('$%s$' %a , xy = (2,0.6),size = 15 ).set_color('black')


ax3=plt.subplot(1,3,3)
ax3.set_xlim(left= 0)
ax3.set_ylim(bottom = 0)
ax3.set_xlim(0,5)
ax3.plot(x, y)
ax3.scatter(x2,y2, zorder = 5)
pole2=0
for i in range(0, ilosc):
    ax3.add_patch(Rectangle((x1[i], 0), x1[1], y1[i+1], facecolor="lightblue", edgecolor="darkblue"))
    pole2= pole2 + y1[i+1]*x1[1]
plt.title(round(pole2,3))

plt.savefig('05_Tab_272348.pdf')
plt.show()
```
