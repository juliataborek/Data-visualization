```python
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
import pandas as pd
import matplotlib.animation as animation
```

```python
countries = pd.read_csv("https://raw.githubusercontent.com/dbouquin/IS_608/master/NanosatDB_munging/Countries-Continents.csv")
```

```python
motor = pd.read_csv("Motor vehicles per 1000 people.csv")
```

```python
l = []
for i in range(len(motor)):
    l.append("a")
```

```python
for i in range(len(motor)):
    Country = motor.iloc[i,0]
    for j in range(len(countries)):
        if Country == countries.iloc[j,1]:
            l[i] = countries.iloc[j,0]
```

```python
motor["Continent"] = l
```
